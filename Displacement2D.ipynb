```python
from scipy.interpolate import interp1d
font = {'size'   : 16, 'family':'STIXGeneral'}
axislabelfontsize='large'
matplotlib.rc('font', **font)
matplotlib.mathtext.rcParams['legend.fontsize']='medium'
```

```python
N = 1200
jump = 120
Box = 30000.0

fac = (2.*pi/Box)**(1.5)
kgauge = 5e-3
kpeak = 0.02
c = 3*kpeak**4
kmax = N*pi/Box
print 'kmin = ',2*pi/Box,'kmax = ',kmax

use_useless_formula = False
if use_useless_formula:
    def pk1fun(k):
        return (k/(k**4+c))
    def pk2fun(k):
        return (k/(k**4+c) + (kgauge/k)**3)
else:
    tmp = loadtxt('output/disp00_z2_matterpk.dat')
    pk1fun = interp1d(tmp[:,0],tmp[:,1])
    kmax = min(kmax,tmp[-1,0])
    tmp = loadtxt('output/disp00_z2_disp_boostpk.dat')
    pk2fun = interp1d(tmp[:,0],tmp[:,1])
    kmax = min(kmax,tmp[-1,0])

RR = numpy.random.normal(loc=0.0, scale=1.0, size=(N,N))
xf1 = np.zeros((N,N),dtype=complex)
xf2 = np.zeros((N,N),dtype=complex)


kvec = [0,0]
for kx in range(N):
    kxm = (N - kx)%N
    for ky in range(N/2+1):
        kym = (N-ky)%N
        if (kx==0) and (ky==0):
            xf1[0,0] = 0.
            continue
        if (ky==kym) and (kx>N/2):
            continue
        if (kx<N/2):
            kvec[0] = kx*2*pi/Box
        else:
            kvec[0] = -(N-kx)*2*pi/Box
        if (ky<N/2):
            kvec[1] = ky*2*pi/Box
        else:
            kvec[1] = -(N-ky)*2*pi/Box
        
        kmag2 = kvec[0]*kvec[0]+kvec[1]*kvec[1]
        kmag = sqrt(kmag2)
        if kmag>kmax:
            continue

        #Pk1 = kmag/(kmag2*kmag2+c)
        #Pk2 = Pk1 + (kgauge/kmag)**3     
        #t1 = sqrt(Pk1)/kmag2
        #t2 = sqrt(Pk2)/kmag2        
        #if 0==1:
        #    t1 *=kmag2
        #    t2 *=kmag2
        
        t1 = fac*sqrt(pk1fun(kmag))/kmag2
        t2 = fac*sqrt(pk2fun(kmag))/kmag2
        
        if kx==kxm and ky==kym:
            xf1[kx,ky] = t1*RR[kx,ky]
            xf2[kx,ky] = t2*RR[kx,ky]
        else:
            xf1[kx, ky] = t1*(RR[kx,ky]+1j*RR[kxm,kym])
            xf1[kxm, kym] = t1*(RR[kx,ky]-1j*RR[kxm,kym])
            xf2[kx, ky] = t2*(RR[kx,ky]+1j*RR[kxm,kym])
            xf2[kxm, kym] = t2*(RR[kx,ky]-1j*RR[kxm,kym])
```
