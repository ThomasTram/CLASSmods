```python
import numpy as np
from classy import Class
from numpy.polynomial.legendre import legval
```

```python
cosmo = Class()
cosmo.set({'output':'lCl, tCl','lensing':'yes'})
cosmo.compute()
rcl = cosmo.raw_cl()
lcl = cosmo.lensed_cl()
```

```python
theta = logspace(-5,log10(pi),1000)
rCtheta = legval(cos(theta),(2.*rcl['ell']+1.)*rcl['tt']/(4.*np.pi))
lCtheta = legval(cos(theta),(2.*lcl['ell']+1.)*lcl['tt']/(4.*np.pi))

semilogx(theta*180./np.pi,rCtheta)
semilogx(theta*180./np.pi,lCtheta,'--')
#ylim([-1e-10,5e-10])
xlabel(r'$\theta \, [\mathrm{deg}]$')
ylabel(r'$C(\theta)^{TT}$')
legend(['Unlensed','Lensed'])
savefig('Cltheta.png')
```
