```python
from classy import Class
```

```python
cosmovec = []
zvec = [0.0, 6.0, 49.0, 1000.0]
zvecstring = str(zvec)[1:-1]
print zvecstring

cosmovec.append(Class())
cosmovec[0].set({'output':'tCl lCl mPk dTk vTk','z_pk':zvecstring,'gauge':'Syn','N_ncdm':1,'N_ur':2,'m_ncdm':1.0})
cosmovec[0].compute()
cosmovec.append(Class())
cosmovec[1].set({'output':'tCl lCl mPk dTk vTk','z_pk':zvecstring,'gauge':'Syn','N_ur':3.0})
cosmovec[1].compute()
```

```python
leg = []
lsvec = ['-','--',':','-.']

for i in range(len(cosmovec)):
    cosmo = cosmovec[i]
    for j in range(len(zvec)):
        z = zvec[j]
        a = 1.0/(1.0+z)
        H = cosmo.Hubble(z)
        
        ls = lsvec[j]
        tkz = cosmo.get_transfer(z)
        if i==0 and j==0:
            print tkz.viewkeys()
        tkzm = cosmo.get_transfer(z-DZ)
        tkzp = cosmo.get_transfer(z+DZ)
        
        kh = array(tkz['k (h/Mpc)'])
        k = kh*cosmo.h()
        
        dthetadz = (array(tkzp['t_tot'])-array(tkzm['t_tot']))/(2*DZ)
        
        thetadot = -H*dthetadz
        
        
        #Assuming synchronous gauge:
        xi = -1.0/k**2*(a*H*array(tkz['t_tot'])+0.*thetadot)
        #xi = tkz['t_tot']
        loglog(kh,abs(xi),ls=ls,color='r')
        leg.append(r'$|\xi(k, z='+str(z)+')|$')
        loglog(kh,tkz['phi'],ls=ls,color='b')
        leg.append(r'$\Phi(k, z='+str(z)+')$')
        loglog(kh,abs(tkz['xi']),ls=ls,color='g')
        leg.append(r'$\xi^{II}(k, z='+str(z)+')$')


lgd = legend(leg,bbox_to_anchor=(1.5, 1.0))
```
