```python
from classy import Class
```

```python
k_values = [1e-4, 1e-3, 1e-2, 1e-1, 1.0]
v = str(k_values)
v = v[1:-1]
print v
cosmo = Class()
cosmo.set({'gauge':'Newtonian','output':'mPk','k_output_values':v})
cosmo.compute()
```

```python
pt = cosmo.get_perturbations()
```

```python
pts = pt['scalar']
ptsk = pts[0]
ptsk.viewkeys()
```

```python
leg = [];
for i in range(len(pts)):
    ptsk = pts[i]
    loglog(ptsk['a'],ptsk['phi'])
    leg.append(r'$\phi$, $k = '+str(k_values[i])+'$')
    loglog(ptsk['a'],ptsk['R'],ls='--')
    leg.append(r'$\zeta$, $k = '+str(k_values[i])+'$')
leg = legend(leg,ncol=1,loc='center right',bbox_to_anchor=(1.4,0.5))
xlabel(r'$a$')
savefig('phizeta.pdf', bbox_extra_artists=(leg,), bbox_inches='tight')
```

```python
leg = [];
for i in range(len(pts)):
    ptsk = pts[i]
    semilogx(ptsk['a'],ptsk['pzratio'])
    leg.append(r'$k = '+str(k_values[i])+'$')
leg = legend(leg,ncol=1,loc='upper right')
xlabel(r'$a$')
ylabel(r'$\phi/\zeta-3/5$')
ylim([-0.2,0.2])
xlim([1e-4,1])
axhline(y=0,ls='--',color='black')
savefig('phizetaratio.pdf', bbox_extra_artists=(leg,), bbox_inches='tight')
```
