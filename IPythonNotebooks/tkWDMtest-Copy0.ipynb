```python
kmax = 30.0
sampling = 10.
tcgoback = 0.1

#prec = {'P_k_max_1/Mpc':kmax,'ncdm_fluid_approximation':3,'k_per_decade_for_pk':200.,
#        'hyper_sampling_flat':50.,'hyper_sampling_curved_high_nu':50,
#        'hyper_sampling_curved_low_nu':50,'tol_ncdm_bg':1e-10,'tol_ncdm_newtonian':1e-12,
#        'perturb_sampling_stepsize':0.01,
#        'k_max_tau0_over_l_max':5.0,
#        'start_small_k_at_tau_c_over_tau_h':0.0015/tcgoback,
#        'start_large_k_at_tau_h_over_tau_k':0.07/tcgoback,
#        'tight_coupling_trigger_tau_c_over_tau_h':0.015/tcgoback,
#        'tight_coupling_trigger_tau_c_over_tau_k':0.01/tcgoback,
#        'start_sources_at_tau_c_over_tau_h':0.008/tcgoback}

#prec = {'P_k_max_1/Mpc':kmax,'ncdm_fluid_approximation':3,'k_per_decade_for_pk':200.,
#        'tol_ncdm_newtonian':1e-12,'tol_ncdm_synchronous':1e-12,
#        'start_small_k_at_tau_c_over_tau_h':0.0015/tcgoback,
#        'start_large_k_at_tau_h_over_tau_k':0.07/tcgoback,
#        'tight_coupling_trigger_tau_c_over_tau_h':0.015/tcgoback,
#        'tight_coupling_trigger_tau_c_over_tau_k':0.01/tcgoback,
#        'start_sources_at_tau_c_over_tau_h':0.008/tcgoback,
#        'k_step_sub':0.05/sampling,
#        'k_step_super':0.002/sampling,'k_step_transition':0.2/sampling}

prec = {'P_k_max_1/Mpc':kmax,
        'ncdm_fluid_approximation':3,
#        'l_max_g':100,
#        'l_max_ncdm':100,
#        'k_per_decade_for_pk':200.,
        'tol_ncdm_newtonian':1e-9,
#        'tol_ncdm_synchronous':1e-8,
#        'tight_coupling_trigger_tau_c_over_tau_k':0.01/tcgoback,
#        'start_large_k_at_tau_h_over_tau_k':0.07/tcgoback,
#        'tol_perturb_integration':1e-10
        }

WDM = Class()
WDM.set({'N_ur':0.,'YHe':0.25,'N_ncdm':1,'m_ncdm':1e3,'Omega_cdm':0.00000,'Omega_ncdm':0.25,
         'gauge':'Newtonian','output':'mPk,dTk','background_verbose':1,'perturbations_verbose':1,
#         'k_output_values':'1.0,1.1,1.2,1.3,5.0,5.01,5.02,5.03,10.0,10.01,10.02,10.03,30.0,30.01,30.02,30.03'})
#         'k_output_values':'0.1,0.11,0.12,0.13,1.0,1.1,1.2,1.3,5.0,5.01,5.02,5.03,10.0,10.01,10.02,10.03'})
         'k_output_values':'0.1,0.11,0.12,0.13,5.0,5.01,5.02,5.03,7.0,7.1,7.2,7.3,10.0,10.01,10.02,10.03'})
WDM.set(prec)
WDM.compute()

CDM=Class()
CDM.set({'N_ur':0.,'YHe':0.25,'Omega_cdm':0.25,'P_k_max_1/Mpc':2.*kmax,
         'gauge':'Newtonian','output':'mPk,dTk','k_output_values':WDM.pars['k_output_values']})
CDM.compute()
```

```python
tk = WDM.get_transfer()
pm = WDM.get_primordial()
ba = WDM.get_background()
pt = WDM.get_perturbations()
WDM.struct_cleanup()

tkCDM=CDM.get_transfer()
ptCDM = CDM.get_perturbations()
CDM.struct_cleanup()
```

```python
leg = []
for key in tk:
    if key in ['k (h/Mpc)']:
        continue
    loglog(tk['k (h/Mpc)'],tk[key]**2)
    leg.append(key)

for key in tkCDM:
    if key in ['k (h/Mpc)']:
        continue
    loglog(tkCDM['k (h/Mpc)'],tkCDM[key]**2)
    leg.append(key)
    
legend(leg,loc=2)
```

```python
keq = 9.8e-3
k = tk['k (h/Mpc)']
T_cdm = 45.0/4.0*(keq/k)**2*(0.577216-7./2.+log(4*sqrt(2./3.)*k/keq))

T_cdm_int = interpolate.interp1d(tkCDM['k (h/Mpc)'], tkCDM['d_cdm'])
#print k
#print tkCDM['k (h/Mpc)']
T_cdm = T_cdm_int(k)
```

```python
leg = []
for key in tk:
    if key in ['k (h/Mpc)','d_g']:
        continue
    loglog(tk['k (h/Mpc)']/WDM.h(),abs(tk[key]/T_cdm))
    leg.append(key)
    

legend(leg,loc=3)
xlim([0.1,40])
```
