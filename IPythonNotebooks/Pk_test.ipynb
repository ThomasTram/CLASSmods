```python
from classy import Class
```

```python
kmax = 10.
kvec = logspace(-3,log10(kmax),200)
p1 = {'output':'mPk','P_k_max_1/Mpc':kmax,'Omega_cdm':0.25,'Omega_b':0.05,'YHe':0.26}
cosmo1 = Class()
cosmo1.set(p1)
cosmo1.compute()

p2 = {'output':'mPk','P_k_max_1/Mpc':kmax,'Omega_cdm':0.29,'Omega_b':0.01,'YHe':0.26}
cosmo2 = Class()
cosmo2.set(p2)
cosmo2.compute()

p3 = {'output':'mPk','P_k_max_1/Mpc':kmax,'Omega_cdm':0.20,'Omega_b':0.10,'YHe':0.26}
cosmo3 = Class()
cosmo3.set(p3)
cosmo3.compute()

pk1 = []
pk2 = []
pk3 = []
for k in kvec:
    pk1.append(cosmo1.pk(k,0.))
    pk2.append(cosmo2.pk(k,0.))
    pk3.append(cosmo3.pk(k,0.))
    
cosmo1.struct_cleanup()
cosmo2.struct_cleanup()
cosmo3.struct_cleanup()
```

```python
loglog(kvec,pk1,kvec,pk2,kvec,pk3)
```
