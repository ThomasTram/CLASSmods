```python
from classy import Class
```

```python
kmax = 10.0
tcgoback = 10.

CDM = Class()
CDM.set({'N_ur':0.,'YHe':0.25,'Omega_cdm':0.25,'output':'mPk,dTk','P_k_max_1/Mpc':kmax})
CDM.compute()

sampling = 5.

prec = {'N_ur':0.,'YHe':0.25,'N_ncdm':1,'m_ncdm':1e2,'Omega_cdm':0.01,'Omega_ncdm':0.24,
        'P_k_max_1/Mpc':kmax,'ncdm_fluid_approximation':3,'tol_ncdm_bg':1e-10,'tol_ncdm_bg':1e-10,
        'l_max_ncdm':50,'hyper_sampling_flat':30.,'tol_perturb_integration':1.e-6,'perturb_sampling_stepsize':0.01,
        'k_per_decade_for_pk':50.,'k_step_sub':0.05/sampling,'k_step_super':0.002/sampling,'k_step_transition':0.2/sampling}

WDMs = Class()
WDMs.set({'N_ur':0.,'YHe':0.25,'N_ncdm':1,'m_ncdm':1e2,'Omega_cdm':0.01,'Omega_ncdm':0.24,'output':'mPk , dTk'})
WDMs.set(prec)
WDMs.compute()

WDMs2 = Class()
WDMs2.set({'N_ur':0.,'YHe':0.25,'N_ncdm':1,'m_ncdm':1e3,'Omega_cdm':0.01,'Omega_ncdm':0.24,'output':'mPk , dTk'})
WDMs2.set(prec)
WDMs2.compute()

WDMb = Class()
WDMb.set({'N_ur':0.,'YHe':0.25,'Omega_b':0.01,'N_ncdm':1,'m_ncdm':1e2,'Omega_cdm':0.01,'Omega_ncdm':0.24,'output':'mPk,dTk'})
WDMb.set(prec)
WDMb.compute()

WDMn = Class()
WDMn.set({'N_ur':0.,'YHe':0.25,'N_ncdm':1,'m_ncdm':1e2,'Omega_cdm':0.0,'Omega_ncdm':0.24,'output':'mPk,dTk','gauge':'Newtonian'})
WDMn.set(prec)
WDMn.compute()

kvec = logspace(-3,log10(kmax),1000)
WDMs_pk = []
WDMb_pk = []

WDMn_pk = []
CDM_pk = []

for k in kvec:
    WDMs_pk.append(WDMs.pk(k,0.))
    WDMn_pk.append(WDMn.pk(k,0.))
    WDMb_pk.append(WDMb.pk(k,0.))
    CDM_pk.append(CDM.pk(k,0.))

#loglog(kvec,WDMs_pk,kvec,WDMn_pk,kvec,CDM_pk)
loglog(kvec,WDMs_pk,kvec,WDMb_pk,kvec,WDMn_pk)
legend(['s','b','n'])



```

```python
dTkWDMs = WDMs.get_transfer()
#dTkWDMs['ad'].viewkeys()
print dTkWDMs.viewkeys()

dTkWDMs2 = WDMs2.get_transfer()

leg = []
for key in dTkWDMs:
    if key in ['k (h/Mpc)']:
        continue
    loglog(dTkWDMs['k (h/Mpc)'],dTkWDMs[key]**2)
    leg.append(key)

for key in dTkWDMs2:
    if key in ['k (h/Mpc)']:
        continue
    loglog(dTkWDMs2['k (h/Mpc)'],dTkWDMs2[key]**2)
    leg.append(key)
    
legend(leg,loc=2)
```

```python
dTkCDM = CDM.get_transfer()
leg = []
for key in dTkCDM:
    if key in ['k (h/Mpc)']:
        continue
    loglog(dTkCDM['k (h/Mpc)'],dTkCDM[key]**2)
    leg.append(key)

legend(leg,loc=2)
```
